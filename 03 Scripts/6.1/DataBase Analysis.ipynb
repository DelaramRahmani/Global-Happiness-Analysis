```python
import pandas as pd

# Define the correct file path for the dataset
file_path = "/Users/dela/Documents/DataSet happiness reports/DATA/2019.csv"

# Load the dataset
df = pd.read_csv(file_path)

# Display the first few rows to understand the structure
print("First five rows of the dataset:")
display(df.head())

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# Drop missing values if necessary
df = df.dropna()

# Check for duplicate rows
duplicates = df.duplicated().sum()
print("\nNumber of duplicate rows:", duplicates)

# Remove duplicates if any
df = df.drop_duplicates()

# Standardize column names (lowercase & replace spaces)
df.columns = df.columns.str.lower().str.replace(" ", "_")

# Check data types of columns
print("\nData types of each column:")
print(df.dtypes)

# Ensure numerical columns are in correct format
num_columns = ['happiness_score', 'gdp_per_capita', 'social_support', 'life_expectancy']

for col in num_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Display summary statistics
print("\nSummary statistics:")
display(df.describe())

# Save the cleaned dataset
cleaned_file_path = "/Users/dela/Documents/DataSet happiness reports/DATA/cleaned_2019.csv"
df.to_csv(cleaned_file_path, index=False)

print("\n✅ Data Cleaning Complete! Cleaned dataset saved as 'cleaned_2019.csv'.")
```

```python
import pandas as pd

# Define the correct file path for the dataset
file_path = "/Users/dela/Documents/DataSet happiness reports/DATA/2019.csv"

# Load the dataset
df = pd.read_csv(file_path)

# Standardize column names (lowercase, strip spaces, replace spaces with underscores)
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")

# Display updated column names
print("Updated column names:", df.columns)

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# Drop missing values if necessary
df = df.dropna()

# Check for duplicate rows
duplicates = df.duplicated().sum()
print("\nNumber of duplicate rows:", duplicates)

# Remove duplicates if any
df = df.drop_duplicates()

# Correct column references based on actual names
num_columns = ['score', 'gdp_per_capita', 'social_support', 'healthy_life_expectancy']

for col in num_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Display summary statistics
print("\nSummary statistics:")
display(df.describe())

# Save the cleaned dataset
cleaned_file_path = "/Users/dela/Documents/DataSet happiness reports/DATA/cleaned_2019.csv"
df.to_csv(cleaned_file_path, index=False)

print("\n✅ Data Cleaning Complete! Cleaned dataset saved as 'cleaned_2019.csv'.")
```

# Correlations Between Key Variables

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix
correlation_matrix = df.corr()

# Display correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Happiness Factors (2019)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns (excluding 'country_or_region')
numeric_df = df.select_dtypes(include=['number'])

# Compute correlation matrix
correlation_matrix = numeric_df.corr()

# Display correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Happiness Factors (2019)")
plt.show()
```
